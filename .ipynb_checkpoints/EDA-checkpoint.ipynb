## Limpieza de datos

**1.** Crear una copia de los datos base para su manipulación y procesamiento.

```python
telco_data = telco_base_data.copy()
```

**2.** Los cargos totales deben ser una cantidad numérica. Vamos a convertirlo a tipo de dato numérico.

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

**3.** Como podemos ver, faltan 11 valores en la columna TotalCharges. Procedemos a la revisión de estos registros.

```python
telco_data.loc[telco_data ['TotalCharges'].isnull() == True]
```

**4.** Tratamiento de valor perdido.

(11/7043)*100 = 0.15%

Dado que el porcentaje de estos registros en comparación con el conjunto de datos total es muy bajo, es decir, 0,15 %, es seguro ignorarlos para un procesamiento posterior.

```python
#Eliminando valores faltantes
telco_data.dropna(how = 'any', inplace = True)
```

```python
# Chequeo de la estadística descriptiva de variables n\'mericas
telco_data.describe()
```

**5.** Dividir a los clientes en contenedores según la antigüedad, p. para permanencia < 12 meses: asigne un grupo de permanencia si 1-12, para permanencia entre 1 y 2 años, grupo de permanencia de 13-24; y demas...

```python
# Obtener la permanencia máxima.
print(telco_data['tenure'].max())
```

```python
# Agrupar la tenencia en bins de 12 meses
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

**6.** Eliminar las columnas que no se requieren para el procesamiento


```python
#Eliminar columna ID de cliente y tenencia
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telco_data.head()
```

## Exploración de datos

**1.** Trazar distribución de predictores individuales por abandono.

### Univariate Analysis

```python
for i, predictor in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
```

**2.** Convertir la variable de destino 'Churn' en una variable numérica binaria, es decir, Sí = 1; No = 0

```python
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes',1,0)
```

```python
telco_data.head()
```

**3.** Convertir todas las variables categóricas en variables ficticias

```python
# Printing unique values from all categoric columns
def print_unique_col_values(df, dtypes='object'):
    for column in df:
        if df[column].dtypes == dtypes:
            print(f'{column} : {df[column].unique()}')
```

```python
print_unique_col_values(telco_data)
```

```python
telco_data.replace('No internet service', 'No', inplace=True)
telco_data.replace('No phone service', 'No', inplace=True)
```

```python
telco_data.head()
```

```python
print_unique_col_values(telco_data)
```

```python
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup',
                  'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
for column in yes_no_columns:
    telco_data[column].replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
print_unique_col_values(telco_data)
```

```python
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup',
                  'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
for column in yes_no_columns:
    telco_data[column].replace({ 1: True, 0 : False}, inplace=True)
```

```python
telco_data.head()
```

```python
telco_data_dummies = pd.get_dummies(data=telco_data, columns=['InternetService', 'Contract', 'PaymentMethod', 'tenure_group'], dtype=np.uint8)
```

```python
print_unique_col_values(telco_data_dummies)
```

```python
telco_data_dummies['gender'].replace({'Female': 1, 'Male': 0}, inplace=True)
```

```python
telco_data_dummies.head()
```

**4.** Relación entre Cargos Mensuales y Cargos Totales.

```python
sns.lmplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Los cargos totales aumentan a medida que aumentan los cargos mensuales, como se esperaba.

**5.** Churn por Cargos Mensuales y Cargos Totales.

```python
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", fill= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Densidad')
Mth.set_xlabel('Cargos mensuales')
Mth.set_title('Cargos mensuales por abandono')
```

**Perspectiva:** Churn es alto cuando los cargos mensuales son altos.

```python
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", fill= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Densidad')
Tot.set_xlabel('Total de Cargos')
Tot.set_title('Total de cargos por abandono')
```

**Perspectiva sorprendente** como una tasa de abandono más alta con cargos totales más bajos

Sin embargo, si combinamos los conocimientos de 3 parámetros, es decir, la tenencia, los cargos mensuales y los cargos totales, la imagen es un poco clara: un cargo mensual más alto a una tenencia más baja da como resultado un cargo total más bajo. Por lo tanto, todos estos 3 factores, a saber, **Cargo mensual más alto**, **Tiempo más bajo** y **Cargo total más bajo**, están vinculados a **Alto abandono**.

**6. Crear una correlación de todos los predictores con 'Churn'**

```python
plt.figure(figsize=(20,8))
telco_data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

**Información derivada: **

**ALTA** Se observa abandono en caso de **contratos de mes a mes**, **sin seguridad en línea** e **Internet de fibra óptica* *

**BAJO** El churn se ve en el caso de **Contratos a largo plazo**, **Suscripciones sin servicio de Internet** y **Los clientes comprometidos por más de 5 años**

Factores como **Género**, **Disponibilidad del servicio telefónico** y **# de líneas múltiples** tienen casi **NO** impacto en Churn

Esto también es evidente en el **Mapa de calor** a continuación.

```python
plt.figure(figsize=(12,12))
sns.heatmap(telco_data_dummies.corr(), cmap="Paired")
```

### Bivariate Analysis

```python
new_df1_target0=telco_data.loc[telco_data["Churn"]==0]
new_df1_target1=telco_data.loc[telco_data["Churn"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribución de Género para Clientes perdidos',hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod',title='Distribución de PaymentMethod para clientes perdidos',hue='gender')
```

```python
uniplot(new_df1_target1,col='Contract',title='Distribución de contrato para clientes perdidos',hue='gender')
```

```python
uniplot(new_df1_target1,col='TechSupport',title='Distribución de TechSupport para clientes perdidos',hue='gender')
```

```python
uniplot(new_df1_target1,col='SeniorCitizen',title='Distribución de SeniorCitizen para Clientes Churned',hue='gender')
```
